```python
path = 'D:\code-python\project_practice\server_log\kong-logs\kong-logs\logs'
```

```python
pattern = r'(?P<ip>\d+\.\d+\.\d+\.\d+) - - \[(?P<timestamp>.*?)\] "(?P<method>\w+) (?P<endpoint>.*?) HTTP/(?P<http_version>\d+\.\d+)" (?P<status>\d+) (?P<bytes>\d+) "(?P<referrer>.*?)" "(?P<user_agent>.*?)" kong_request_id: "(?P<kong_request_id>.*?)"'
```

```python
log_files = [file for file in os.listdir(path) if file.endswith('000Z')]
```

```python
dataframes = []
```

```python
for filename in log_files:
    file_path = os.path.join(path, filename)
    
    # Đọc file log
    with open(file_path, 'r') as file:
        log_lines = file.readlines()
    
    # Trích xuất dữ liệu từ mỗi dòng log
    log_data = [re.match(pattern, line).groupdict() for line in log_lines if re.match(pattern, line)]
    
    # Chuyển dữ liệu thành DataFrame
    df = pd.DataFrame(log_data)
    df['source_file'] = filename  # Thêm tên file vào mỗi DataFrame
    dataframes.append(df)
```
